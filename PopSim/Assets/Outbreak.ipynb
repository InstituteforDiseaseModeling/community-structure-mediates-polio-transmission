```python
class Results:
    def __init__(self, file):
        match = re.search('summary_(.+)_(.+)_(.+)_(.+).json', file)
        strain, tx_type, year, p = match.groups()
        self.model = tx_type
        self.year = year
        self.origin = strain
        self.p = p
        self.results = json.load(open(file))
        self.peak_time = []
        self.peak_size = []
        
        lengths = [len(repetition) for repetition in self.results['total']]
        max_length = np.max(lengths)
        self.max_length = max_length
        for repetition in self.results['total']:
            while len(repetition) != max_length:
                repetition.append(0)
        self.results['fail'] = []
        for repetition in self.results['total']:
            self.results['fail'].append(np.asarray(repetition[1:]) ==0)
        self.results['t_extinction'] = []
        for repetition in self.results['total']:
            peak_idx = np.argmax(repetition)
            self.peak_time.append(peak_idx)
            self.peak_size.append(repetition[peak_idx])
            if 0 in repetition[1:]:
                self.results['t_extinction'].append(repetition[1:].index(0) + 1)
            else:
                self.results['t_extinction'].append(None)
        
    def plot_curves(self, color):
        plt.plot([x for x in range(self.max_length)], 
                 np.mean(self.results['total'], axis = 0), color = color, label = self.year)
        min_range = np.mean(self.results['total'], axis = 0) - 1.96*np.std(self.results['total'], axis = 0)
        min_range =[x if x > 0 else 0 for x in min_range]
        max_range = np.mean(self.results['total'], axis = 0) + 1.96*np.std(self.results['total'], axis = 0)
        plt.fill_between([x for x in range(self.max_length)], min_range, max_range, color = color, alpha = 0.3)
        
    def plot_log_curves(self, color):
        self.log_results = np.log10(self.results['total'])
        plt.plot([x for x in range(self.max_length)], 
                 np.mean(self.log_results['total'], axis = 0), color = color, label = self.year)
        min_range = np.mean(self.log_results['total'], axis = 0) - 1.96*np.std(self.log_results['total'], axis = 0)
        min_range =[x if x > 0 else 0 for x in min_range]
        max_range = np.mean(self.log_results['total'], axis = 0) + 1.96*np.std(self.log_results['total'], axis = 0)
        plt.fill_between([x for x in range(self.max_length)], min_range, max_range, color = color, alpha = 0.3)

        
def read_data(strain, suffix = 'outbreak'):
    multiscale_results, global_results = {}, {}
    c_multiscale_results, c_global_results = {},{}
    for file in glob.glob('/Users/weswong/Documents/GitHub/MultiscaleModeling/PopSim/outbreak_paper_sims/summary_{suffix}_{strain}_*json'.format(suffix = suffix, strain=strain)):
        print(file)
        match = re.search('summary_(.+)_(.+)_(.+)_(.+).json', file)
        strain, tx_type, year, p = match.groups()
            
        year = int(year)
        p = float(p)
        param = str(year) + '_' + str(p)
        if p == 0.0:
            if tx_type == 'multi':
                multiscale_results[year]= Results(file)
            else:
                global_results[year] = Results(file)
        else:
            if tx_type == 'multi':
                c_multiscale_results[param] = Results(file)
            else:
                c_global_results[param] = Results(file)
    return multiscale_results, global_results, c_multiscale_results, c_global_results

S2_standard = read_data('S2')

def read_data_wpv(strain, suffix = 'outbreak'):
    multiscale_results, global_results = {}, {}
    c_multiscale_results, c_global_results = {},{}
    for file in glob.glob('/Users/weswong/Documents/GitHub/MultiscaleModeling/PopSim/outbreak_paper_sims/summary_{strain}_*json'.format(suffix = suffix, strain=strain)):
        print(file)
        match = re.search('summary_(.+)_(.+)_(.+)_(.+).json', file)
        strain, tx_type, year, p = match.groups()
            
        year = int(year)
        p = float(p)
        param = str(year) + '_' + str(p)
        if p == 0.0:
            if tx_type == 'multi':
                multiscale_results[year]= Results(file)
            else:
                global_results[year] = Results(file)
        else:
            if tx_type == 'multi':
                c_multiscale_results[param] = Results(file)
            else:
                c_global_results[param] = Results(file)
    return multiscale_results, global_results, c_multiscale_results, c_global_results


WPV_standard = read_data_wpv('WPV')
```
