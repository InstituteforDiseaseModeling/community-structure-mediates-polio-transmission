```python
def get_immunities(sim):
    age_bins = range(0, 85, 1)
    basic_stats ={'total': 0,
                 'under5':0,
                 'susceptible_total':0,
                 'susceptible_under5':0}
    age_bin_distribution = defaultdict(list)
    for V in sim.villages:
        for bari in V.baris.values():
            for hh in bari.households.values():
                for individual in hh.individuals.values():
                    age_idx = int(np.digitize(math.floor(individual.age), age_bins))
                    age_bin_distribution[age_idx].append(np.log(individual.infection.current_immunity) / np.log(2))
                    basic_stats['total'] +=1
                    if individual.infection.current_immunity < 8:
                        basic_stats['susceptible_total'] += 1
                    if individual.age < 5:
                        basic_stats['under5'] += 1
                        if individual.infection.current_immunity < 8:
                            basic_stats['susceptible_under5'] += 1
                        
    summary_age_bin = {}
    for age_idx in age_bin_distribution:
        summary_age_bin[age_idx] = [np.median(age_bin_distribution[age_idx]), 
                                    np.percentile(age_bin_distribution[age_idx], 5),
                                   np.percentile(age_bin_distribution[age_idx], 95)]
    return summary_age_bin, basic_stats
```

```python
immunity_profiles = defaultdict(list)
simulations = defaultdict(list)
means = defaultdict(lambda: defaultdict(list))
per5s = defaultdict(lambda: defaultdict(list))
per95s = defaultdict(lambda: defaultdict(list))

basic_stats =defaultdict(list)

for file in glob.glob('../Demographics/Outbreak_Villages/*.pkl'):
    match = re.search('../Demographics/Outbreak_Villages/Outbreak_Villages_.+_(.+).pkl', file)
    year= int(match.groups()[-1])
    print(file, year)
    if year in [0,1,5,10,40]:
        data = dill.load(open(file, 'rb'))
        immune_profiles, demo_stats = get_immunities(data)
        for key in demo_stats:
            basic_stats[year][key].append(demo_stats[key])
        
        immunity_profiles[int(year)].append(immune_profiles)
        for iteration in immunity_profiles[year]:
            for age in iteration:
                mean, per5, per95 = iteration[age]
                means[year][age].append(mean)
                per5s[year][age].append(per5)
                per95s[year][age].append(per95)
        
        
        #simulations[0].append(dill.load(open(file, 'rb')))
```

```python
plt.figure(figsize = (15,10))
plt.subplot(2,1,1)
for year in sorted(means):
    age_averages = [np.mean(means[year][age]) for age in range(86)]
    per5 = [np.nanmean(per5s[year][age]) for age in range(86)]
    per95 = [np.nanmean(per95s[year][age]) for age in range(86)]
    plt.plot(range(len(age_averages)), age_averages, label = year)
    plt.fill_between(range(86), per95, per5, alpha = 0.3)
    
plt.xlabel('Age', fontsize = 15)
plt.ylabel('OPV-equivalent Antibody Titer (log2)', fontsize = 15)
plt.xticks(size = 12)
plt.yticks(size = 12)
plt.legend(fontsize = 12)
plt.title('A', loc='left', fontsize = 20)

plt.subplot(2,1,2)
susceptibles = defaultdict(lambda: dict())
for year in [0,1,5,10,40]:
    susceptibles_5=np.asarray(basic_stats[year]['susceptible_under5']) / np.asarray(basic_stats[year]['under5'])
    susceptibles_total = np.asarray(basic_stats[year]['susceptible_total']) / np.asarray(basic_stats[year]['total'])
    susceptibles['under_5'][year] = susceptibles_5
    susceptibles['total'][year] = susceptibles_total
    
    
plt.plot([0,1,5,10,40], [np.mean(susceptibles['under_5'][year]) for year in [0,1,5,10,40]])

plt.scatter([0,1,5,10,40], [np.mean(susceptibles['under_5'][year]) for year in [0,1,5,10,40]])

plt.plot([0,1,5,10,40], [np.mean(susceptibles['total'][year]) for year in [0,1,5,10,40]])

plt.scatter([0,1,5,10,40], [np.mean(susceptibles['total'][year]) for year in [0,1,5,10,40]])

plt.tick_params(labelsize = 12)
plt.xlabel('Years Post-Switch', fontsize = 15)
plt.ylabel('Proportion Susceptible', fontsize = 15)
plt.title('B', loc='left', fontsize = 20)
plt.savefig('/Users/weswong/Dropbox (Personal)/multiscaleTransmissionTeam/Multiscale_Bangladesh_Paper/Publication Figures' + '/SFig_2.png',dpi=300)
```
