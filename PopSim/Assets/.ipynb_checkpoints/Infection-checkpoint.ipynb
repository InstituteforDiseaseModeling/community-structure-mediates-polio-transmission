```python
def reinfection_plots(repetitions, t_before_challenge = 120, reinfection_cycles = 5, color = 'blue', alpha = 1, initial_age = 2):
    data = defaultdict(lambda: np.zeros((repetitions, t_before_challenge * reinfection_cycles + 1)))
    for rep in range(repetitions):
        I= ImmunoInfection('S2')
        p_shed = []
        p_infection = []
        current_immunity = []
        shed_virions = []
        theta = []
        
        p_shed.append(0)
        p_infection.append(1)
        current_immunity.append(0)
        shed_virions.append(0)
        theta.append(I.theta_Nab())
        
        t = 0
        
        for cycle in range(reinfection_cycles):
            I.attempt_infection()
            for tstep in range(0,t_before_challenge):
                p_shed.append(I.is_shed)
                p_infection.append(I.p_infection())
                current_immunity.append(I.current_immunity)
                current_age = initial_age + t/365.
                shed_virions.append(I.viral_shed(initial_age + (cycle + 1)*(tstep/365.)))
                theta.append(I.theta_Nab())
                t += 1
                I.update(1/365.)

        
        data['p_shed'][rep] = p_shed
        data['p_infection'][rep] = p_infection
        data['current_immunity'][rep] = current_immunity
        data['shed_virions'][rep] = shed_virions
        data['theta'][rep] = np.log2(theta)
    return data
```

```python
t_before_challenge, reinfection_cycles = 720,5
infection_data = reinfection_plots(1000, t_before_challenge, reinfection_cycles)
```

```python
def normalize(array):
    return np.asarray(array) / np.sum(array)


def infection_rate(x, alpha, beta = 23, gamma =2):
    rate = beta * (1 - np.exp(-alpha*x) ) + gamma
    if rate < 0:
        print('out of bounds')
        print(alpha, beta, gamma)
    return rate

class Param_Search:
    def __init__(self):
        pass
            
    def simulate(self,params):
        alpha, beta, gamma = params
        
    def taniuchi_decay_immunity(self, age, initial_Nab = 580, rate=0.24):
        '''Average empirical decay in immunity observed in household contacts > 5 years from Taniuchi et al (2018)
        age in years'''
        return initial_Nab * (1+ (12.*age - 30)) ** - rate

        
    def reinfection_model(self, alpha, beta, gamma):
        n_months = 80 * 12
        I= ImmunoInfection('S2')
        current_immunity, current_ages = [], []                
        t_wait_time = np.random.gamma(shape = infection_rate(0, alpha, beta, gamma), scale = 1)
        #1 event every shape months
        
        for i, t_step in enumerate(range(0, n_months)):
            current_age = 0 + (i+1) * 30/365
            if current_age > 5:
                current_ages.append(current_age)
                current_immunity.append(I.current_immunity)
            I.update(30/365) #advance thirty days
            t_wait_time -= 1
            if t_wait_time < 0:
                I.attempt_infection()
                t_wait_time = np.random.gamma(shape = infection_rate(current_age, alpha, beta, gamma), scale = 1)
        return current_ages, current_immunity
    
    def calculate_difference(self, params):
        sim_ages, sim_immunities = self.reinfection_model(*params)
        taniuchi_immunities = [self.taniuchi_decay_immunity(sim_age) for sim_age in sim_ages]
        
        distance = np.sum((np.log2(taniuchi_immunities) - np.log2(sim_immunities))**2)
        return distance
    
    def calculate_distance_distribution(self, priors):
        '''generates a distance distribution (KL divergence technically) and determines the 30th, 10th, 5th percentiles'''
        distances = []
        for _ in range(100):
            if _ %100 == 0:
                print('calculating distance distribution iteration {iteration} / 1000'.format(iteration = _))
            distance = self.calculate_difference([priors[key]() for key in priors])
            distances.append(distance)
        plt.hist(distances)
        percentiles = [np.percentile(distances, 30), np.percentile(distances, 10), np.percentile(distances, 5)]
        return percentiles
    
    def sequential_pmc_ABC(self, n_particles=300, eps = None): 
            '''Population Montecarlo PMC algorithm designed to perform ABC and iteratively adapt the prior/proposal distribution to more closely resemble the posterior
            does this by iteratively reducing the tolerated error and progressively adapting the prior distribution
            follows the Algorithm 2 of Sadegh and vrugt 2014: Approximate Bayesian computation using Markov Chain Monte Carlo simulation: DREAM_ABC_ and the code at https://stackoverflow.com/questions/37496969/population-monte-carlo-implementation
            eps = initial error terms'''
            coefficients_mean = [0.07, 25, 2]
            scales_std = [0.01, 1, 0.2]
            priors = {}
            alphabet_string='abcdefghijklmnopqrstuvwxyz'
            priors['a'] = lambda: np.random.normal(coefficients_mean[0], scales_std[0])
            priors['b'] = lambda: np.random.normal(coefficients_mean[1], scales_std[1])
            priors['c'] = lambda: np.random.normal(coefficients_mean[2], scales_std[2])
            #print([priors[key]() for key in priors])


            def computeWeight(prevWeights,prevParticles,prevVariance,currentParticle):
                '''compute the weights needed to adapt the proposal distribution to be closer to the posterior'''
                prior = np.prod([scipy.stats.norm.pdf(currentParticle, coefficients_mean[0], scales_std[0]), 
                                 scipy.stats.norm.pdf(currentParticle, coefficients_mean[1], scales_std[1]),
                                 scipy.stats.norm.pdf(currentParticle, coefficients_mean[2], scales_std[2])])

                denom = 0.0
                scale = 1e5 
                for i in range(len(prevParticles)):
                    denom += prevWeights[i]  * scipy.stats.multivariate_normal(scale * currentParticle, 
                                                                               scale * prevVariance,
                                                                              allow_singular=True).pdf(scale * prevParticles[i])
                return prior/denom    

            if not eps:
                eps = self.calculate_distance_distribution(priors) #automatically calculate the 30th, 10th, 5th percentiles if not given
                print(eps)

            iterations = len(eps)
            distance = 1e9
            variance = {} #for multivariate, this is covariance matrix
            particles = defaultdict(list)
            weights = np.empty([iterations,n_particles])
            simData = np.empty(n_particles)

            for j in range(iterations):
                print('starting {j}th iteration'.format(j=j))
                if j == 0:
                    for i in range(n_particles):
                        while(distance > eps[j]):
                            draw = [priors[key]() for key in priors]
                            while sum(np.asarray(draw) < 0) > 0:
                                draw = [priors[key]() for key in priors]
                            distance = self.calculate_difference(draw)
                        if i in [round(x,0) for x in np.linspace(0, n_particles, 10)]:
                            print(draw, distance)
                            print('particle {i} solution found'.format(i = i))
                        particles[j].append(draw)
                        weights[j][i] = 1./n_particles
                        distance = 1e9
                    particles[j] = np.asarray(particles[j])
                    variance[j] = 2 * np.cov(particles[j].T) #np.var(particles[j].T)
                else:
                    for i in range(n_particles):
                        while(distance > eps[j]):
                            prelim_draw_index = np.random.choice(range(len(particles[j-1])), p = weights[j-1]) #draw a new parameter from the weighted distribution of previously identified parameters; random choice only works on 1-D arrays (no nested)
                            prelim_draw = particles[j-1][prelim_draw_index]
                            #draw = np.random.normal(prelim_draw, np.sqrt(variance[j-1])) #peturb the new prior in proportion to the stdev of the previous iteration
                            #distance = self.calculate_difference([draw])
                            draw = np.random.multivariate_normal(prelim_draw, variance[j-1]) #multivariate_normal accepts the covariance
                            while  sum(np.asarray(draw) < 0) > 0:
                                draw = np.random.multivariate_normal(prelim_draw, variance[j-1])
                            distance = self.calculate_difference(draw)
                        if i in [round(x,0) for x in np.linspace(0, n_particles, 10)]:
                            print('particle {i} solution found'.format(i = i))
                            print(draw, distance)

                        particles[j].append(draw)
                        weights[j][i] = computeWeight(weights[j-1],particles[j-1],variance[j-1],particles[j][i])
                        distance = 1e9
                    particles[j] = np.asarray(particles[j])
                    variance[j] = 2 * np.cov(particles[j].T) #np.var(particles[j].T)
                    weights[j] = normalize(weights[j])
            self.posteriors = np.asarray(particles)
            self.max_posteriors = [np.mean(distribution) for distribution in self.posteriors.item()[j].T] 
        
```

```python
def p_infection(dose = 10**6, strain = 'S2', alpha = 0.44, gamma = 0.46): #dose values are for a typical vaccine dose
        '''Dose response model. Pr(infection | dose, immunity). 
        Immunity is an abstraction of Nab * theta 
        Immunity based on probability of infection'''
        sabin_scale_parameters={'S1': 14, 'S2': 8, 'S3': 18, 'WPV': 2.3}
        beta = sabin_scale_parameters[strain]
        return (1 - (1 + dose/beta)**(-alpha * (1)**-gamma)) #* 0.554
```

```python
class Immunization_Param_Search:
    def __init__(self, trial):
        self.trial = trial
                    
    def bOPV(self, modifier):
        '''Average empirical decay in immunity observed in household contacts > 5 years from Taniuchi et al (2018)
        age in years'''
        repetitions = 200
        shed_prevalences = np.zeros((repetitions, 5))
        for rep in range(repetitions):
            I = ImmunoInfection('S1')
            is_shed = []
            for day in range(18*7):
                I.update(1/365.)
                if day in 7*np.asarray([6,10,14]):
                    I.attempt_infection(modifier = modifier)

            I.attempt_infection()
            for day in range(5*7+1):
                I.update(1/365.)
                if day in [7,14,21,28,35]:
                    is_shed.append(I.is_shed)
            shed_prevalences[rep] = is_shed
        return np.mean(shed_prevalences, axis = 0)
    
    
    def tOPV(self, modifier):
        repetitions = 200
        shed_prevalences = np.zeros((repetitions, 5))
        for rep in range(repetitions):
            I = ImmunoInfection('S2')
            current_immunity = []
            is_shed = []
            for day in range(18*7):
                I.update(1/365.)
                if day in 7*np.asarray([6,10,14]):
                    I.attempt_infection(modifier = modifier)

            I.attempt_infection()
            for day in range(5*7 + 1):
                I.update(1/365.)
                if day in [7,14,21,28,35]:
                    is_shed.append(I.is_shed)
            shed_prevalences[rep] = is_shed
        return np.mean(shed_prevalences, axis = 0)

    def calculate_difference(self, param):
        if self.trial == 'tOPV':
            shed_prevalences = self.tOPV(param)
            study_prevalences = np.asarray([0.35443037974683544, 0.0759493670886076, 0.02531645569620253, 0.012658227848101266, 0.0])
        else:
            shed_prevalences = self.bOPV(param)
            study_prevalences = np.asarray([0.73096447, 0.39593909, 0.34010152, 0.2284264 , 0.11167513])
        
        distance = np.sum((shed_prevalences - study_prevalences)**2)
        return distance
    
    def calculate_distance_distribution(self, priors):
        '''generates a distance distribution (KL divergence technically) and determines the 30th, 10th, 5th percentiles'''
        distances = []
        for _ in range(100):
            if _ %100 == 0:
                print('calculating distance distribution iteration {iteration} / 1000'.format(iteration = _))
            distance = self.calculate_difference(priors['a']())
            distances.append(distance)
        plt.hist(distances)
        percentiles = [np.percentile(distances, 30), np.percentile(distances, 10), np.percentile(distances, 5)]
        return percentiles
    
    def sequential_pmc_ABC(self, n_particles=300, eps = None): 
            '''Population Montecarlo PMC algorithm designed to perform ABC and iteratively adapt the prior/proposal distribution to more closely resemble the posterior
            does this by iteratively reducing the tolerated error and progressively adapting the prior distribution
            follows the Algorithm 2 of Sadegh and vrugt 2014: Approximate Bayesian computation using Markov Chain Monte Carlo simulation: DREAM_ABC_ and the code at https://stackoverflow.com/questions/37496969/population-monte-carlo-implementation
            eps = initial error terms'''
            priors = {}
            if self.trial == 'bOPV':
                priors['a'] = lambda: np.random.normal(0.166, 0.05)
            else:
                priors['a'] = lambda: np.random.normal(0.63, 0.1)

            def computeWeight(prevWeights,prevParticles,prevVariance,currentParticle):
                '''compute the weights needed to adapt the proposal distribution to be closer to the posterior'''
                if self.trial == 'tOPV':
                    prior = scipy.stats.norm.pdf(currentParticle, 0.63, 0.1)
                else:
                    prior = scipy.stats.norm.pdf(currentParticle, 0.166, 0.05)

                denom = 0.0
                scale = 1e5 
                for i in range(len(prevParticles)):
                    denom += prevWeights[i]  * scipy.stats.multivariate_normal(scale * currentParticle, 
                                                                               scale * prevVariance,
                                                                              allow_singular=True).pdf(scale * prevParticles[i])
                return prior/denom    

            if not eps:
                eps = self.calculate_distance_distribution(priors) #automatically calculate the 30th, 10th, 5th percentiles if not given
                print(eps)

            iterations = len(eps)
            distance = 1e9
            variance = {} #for multivariate, this is covariance matrix
            particles = defaultdict(list)
            weights = np.empty([iterations,n_particles])
            simData = np.empty(n_particles)

            for j in range(iterations):
                print('starting {j}th iteration'.format(j=j))
                if j == 0:
                    for i in range(n_particles):
                        while(distance > eps[j]):
                            draw = priors['a']()
                            while draw < 0 or draw > 1:
                                draw = priors['a']()
                            distance = self.calculate_difference(draw)
                        if i in [round(x,0) for x in np.linspace(0, n_particles, 10)]:
                            print(draw, distance)
                            print('particle {i} solution found'.format(i = i))
                        particles[j].append(draw)
                        weights[j][i] = 1./n_particles
                        distance = 1e9
                    particles[j] = np.asarray(particles[j])
                    variance[j] = 2 * np.cov(particles[j].T) #np.var(particles[j].T)
                else:
                    for i in range(n_particles):
                        while(distance > eps[j]):
                            prelim_draw_index = np.random.choice(range(len(particles[j-1])), p = weights[j-1]) #draw a new parameter from the weighted distribution of previously identified parameters; random choice only works on 1-D arrays (no nested)
                            prelim_draw = particles[j-1][prelim_draw_index]
                            draw = np.random.normal(prelim_draw, np.sqrt(variance[j-1])) #peturb the new prior in proportion to the stdev of the previous iteration
                            while draw < 0 or draw > 1:
                                draw = priors['a']()
                            #draw = np.random.multivariate_normal(prelim_draw, variance[j-1]) #multivariate_normal accepts the covariance
                            #while  sum(np.asarray(draw) < 0) > 0:
                            #    draw = np.random.multivariate_normal(prelim_draw, variance[j-1])
                            distance = self.calculate_difference(draw)
                        if i in [round(x,0) for x in np.linspace(0, n_particles, 10)]:
                            print('particle {i} solution found'.format(i = i))
                            print(draw, distance)

                        particles[j].append(draw)
                        weights[j][i] = computeWeight(weights[j-1],particles[j-1],variance[j-1],particles[j][i])
                        distance = 1e9
                    particles[j] = np.asarray(particles[j])
                    variance[j] = 2 * np.cov(particles[j].T) #np.var(particles[j].T)
                    weights[j] = normalize(weights[j])
            self.posteriors = np.asarray(particles)
            self.max_posteriors = [np.mean(distribution) for distribution in self.posteriors.item()[j].T] 
        
```

```python
def bOPV_reinfection_model(modifier = 0.31):
    #rate is per day!
    I= ImmunoInfection('S1')
    p_infection, current_immunity, is_shed = [], [], []                
    for day in range(18*7):
        I.update(1/365.)
        current_immunity.append(I.current_immunity)
        is_shed.append(I.is_shed)
        p_infection.append(I.p_infection() * modifier)
        if day in 7*np.asarray([6,10,14]):
            I.attempt_infection(modifier = modifier)
    
    I.attempt_infection()
    for day in range(5*7):
        I.update(1/365.)
        current_immunity.append(I.current_immunity)
        is_shed.append(I.is_shed)    
        p_infection.append(I.p_infection() * modifier)
    return np.log2(current_immunity), is_shed
```
