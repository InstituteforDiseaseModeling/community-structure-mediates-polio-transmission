```python
def read_asfr_data(file='../data/Bangladesh_DHSS_age_specific_fertility_2014.txt'): #asfr = age-specific fertility rates 
    fin = open(file)
    asfr_rates = {}
    for x in fin.readlines()[1:]:
        age_bin = x.strip().split()[0]
        asfr_rates[age_bin] = float(x.strip().split()[-1])
    return asfr_rates

def read_birth_interval_data(file = '../data/Bangladesh_birth_intervals_2014.txt'):
    data = [_.strip().split(' ') for _ in open('../data/Bangladesh_birth_intervals_2014.txt').readlines() if '#' not in _][0:-1]
    p_births = {}

    birth_intervals = ['7-17', '18-23', '24-35', '36-47', '48-59', '60+']
    t_birth_intervals = [(float(x.split('-')[0])/12.,float(x.split('-')[1])/12.) for x in birth_intervals[:-1]] + [5.]
    
    for line in data[1:]:
        age_cohort = line[0]
        p_births[line[0]] = np.asarray([float(x)/100. for x in line[1:7]])
        
    return p_births

bangladesh_dhss_birth_interval_dict = read_birth_interval_data()

bangladesh_dhss_asfr_dict = read_asfr_data()
bangladesh_2004_asfr_dict = read_asfr_data(file = '../data/Bangladesh_DHSS_age_specific_fertility_2004.txt')
midpoints = [np.mean([float(x.split('-')[0]), float(x.split('-')[1])]) for x in bangladesh_dhss_asfr_dict.keys()]
```

```python
class Distribution:
    '''wrapper class for various distributions to optimize'''
    def __init__(self, pdf, cdf, integral=None):
        self.pdf = pdf #or pmf
        self.cdf = cdf
        self.integral = integral  

        
def weibull_distribution(params_array, x):
    rate, k = params_array[0], params_array[1]
    pdf = k/rate * (x/rate)**(k-1) * np.e**-(x/rate)**k
    return pdf

def weibull_cdf(params_array, x):
    rate, k = params_array[0], params_array[1]
    cdf = 1 - np.e**-(x/rate) **k
    return cdf
    
def weibull_integral(params_array, lower, upper):
    integral = weibull_cdf(params_array, upper) - weibull_cdf(params_array, lower)
    return integral

#exponentiated weibull distributions -- 3 parameter distribution, when alpha =1 the same as a weibull
def weibull_exponentiated_distribution(params_array,x):
    rate, k, alpha = params_array[0], params_array[1], params_array[2]
    return alpha * (k/rate)*(x/rate)**(k-1)*(1- np.e**-(x/rate)**k)**(alpha -1) * np.e**(-(x/rate)**k)

def weibull_exponentiated_cdf(params_array, x):
    rate, k, alpha = params_array[0], params_array[1], params_array[2]
    return (1 - np.e **-(x/rate) **k)**alpha

def weibull_exponentiated_integral(params_array, lower, upper):
    integral = weibull_exponentiated_cdf(params_array, upper) - weibull_exponentiated_cdf(params_array, lower)
    return integral

#x is number of events/successes that occur
Poisson = Distribution(lambda params_array, x: scipy.stats.poisson.pmf(x,params_array[0]), #params_array = [lambda], n is a dummy variable
                      lambda params_array, x: scipy.stats.poisson.cdf(x, params_array[0]))
Negative_Binomial = Distribution(lambda params_array, x: scipy.stats.nbinom.pmf(x, params_array[0], params_array[1]),
                                 lambda params_array, x: scipy.stats.nbinom.cdf(x, params_array[0], params_array[1]))
Weibull = Distribution(weibull_distribution, weibull_cdf, weibull_integral) #params_array = rate,k
Weibull_exponentiated = Distribution(weibull_exponentiated_distribution, weibull_exponentiated_cdf, weibull_exponentiated_integral) #params_array = rate, k, alpha


def lognormal_cdf(x, mu, sigma):
    cdf = 1/2 + (1/2) * erf((np.log(x) - mu ) / (np.sqrt(2) * sigma))
    return cdf

def lognormal_integral(start, stop, mu, sigma):
    integral = lognormal_cdf(stop, mu, sigma) - lognormal_cdf(start, mu, sigma)
    return integral

def sigmoid(p,x):
    x0,y0,c,k=p
    y = c / (1 + np.exp(-k*(x-x0))) + y0
    return y

def multinomial(p_array, x_array):
    n_factorial = gamma_fn(np.sum(x_array) + 1)
    x_i_factorial_product_sum = 1
    for x_i in x_array:
        x_i_factorial = gamma_fn(x_i + 1)
        x_i_factorial_product_sum *= x_i_factorial
    p_i_product_sum = 1
    for x_i, p_i in zip(x_array, p_array):
        p_i_product_sum *= (p_i**x_i)
    
    pmf = (n_factorial / x_i_factorial_product_sum) * p_i_product_sum
    return pmf

def kl_divergence(pk, qk):
    '''From scipy.stats documentation:
    Calculate the entropy of a distribution for given probability values. (pk vs qk)
    If only probabilities pk are given, the entropy is calculated as S = -sum(pk * log(pk), axis=0).
    If qk is not None, then compute the Kullback-Leibler divergence S = sum(pk * log(pk / qk), axis=0).
    This routine will normalize pk and qk if they donâ€™t sum to 1.
    Assumes that pk[i] corresponds to the probablitity of the ith event and that it is a discrete distribution'''
    return scipy.stats.entropy(pk, qk)

def laplace_smoothing(x_array, pseudocount=1):
    N = np.sum(x_array)
    smoothed = (np.asarray(x_array) + pseudocount) / (N + len(x_array)*pseudocount)
    return smoothed

def read_life_table(file):
    fin = open(file)
    ages = [0]
    males, females = [],[]
    for x in fin.readlines()[1:]:
        age_bins = x.strip().split()[0].split('-')
        if len(age_bins) >1:
            ages.append(np.mean([int(age_bins[0]), int(age_bins[1])]))
        males.append(float(x.strip().split()[1]))
        females.append(float(x.strip().split()[2]))
    ages.append(85.)
    
    return ages, np.asarray(females), np.asarray(males)

def normalize(array):
    return np.asarray(array) / np.sum(array)
    
    
def cdf_married(age, coefficients = [0.20978635, 0.47889714, 0.98557189]):
    def growth_rate(x, p0,  r, k):
        '''population growth logistic equation: Verhulst equation'''
        return (k*p0*np.exp(r*x)) / (k + p0*(np.exp(r*x) - 1))
    
    x = age - 15 #shift the growth_rate equation for better calculation
    y = growth_rate(x, *coefficients)
    return y

flatten = lambda l: [item for sublist in l for item in sublist] #flatten a list of of lists
```

```python
class Individual:
    id_generator = itertools.count()
    def __init__(self, age, base_fertility_params, scale = 1):
        self.id = next(Individual.id_generator) #iterator that generates a new unique id everytime it is called
        self.age = age
        self.base_fertility_params = base_fertility_params
        self.isalive =True
        self.check_fertility()
        if type(scale) == int or type(scale) == float:
            self.scale = lambda x: scale
        else:
            self.scale = scale
        
        
    def check_alive(self, dt): #check if person is still alive after specified time period
        p_death = 1-np.e**(-self.mortality_rate()*dt)
        if np.random.random() < p_death:
            self.isalive = False
            self.death_age = self.age
            
    def check_fertility(self):
        if self.age >=15 and self.age <50: #assume males virility follow fertile rules in females
            self.isfertile = True
        else:
            self.isfertile = False
    
    def polynomial_fertility(self):
        a,b,c = self.base_fertility_params[0], self.base_fertility_params[1], self.base_fertility_params[2]
        if self.age < 15 or self.age > 50:
            return 0
        else:
            return np.max([0, a*self.age**2 + b*self.age + c])
        
    def constant_fertility(self):
        if self.age < 15 or self.age > 50:
            return 0
        else:
            p_child = self.base_fertility_params[0]
            return p_child
        
    def double_exp_fertility(self):
        a,b,c = self.base_fertility_params[0], self.base_fertility_params[1], self.base_fertility_params[2]
        if self.age < 15 or self.age > 50:
            return 0
        else:
            x = self.age
            return a * np.exp(-np.exp(-((x-b)/c))-((x-b)/c))
    
    def poly_4(self):
        a,b,c,d,e = self.base_fertility_params[0], self.base_fertility_params[1], self.base_fertility_params[2],self.base_fertility_params[3],self.base_fertility_params[4]
        if self.age < 15 or self.age > 50:
            return 0
        else:
            x = self.age
            return max(0, a *x**4 + b*x**3 + c*x**2 + d*x + e)
    
    
    def base_p_child(self, dt):
        '''baseline probabilities of having a child-- P(Child | no_children)'''
        #p_child =1-np.e**(-f_rate*dt)
        p_child = self.poly_4() * self.scale(self.age)
        #p_child = self.double_exp_fertility() #self.polynomial_fertility()
        return p_child 
    
    @classmethod
    def create_child(cls):
        if np.random.random() > 0.5:
            return Male(0)
        else: 
            return Female(0)


class Male(Individual):
    def __init__(self, *args, **kwargs):
        Individual.__init__(self, *args, **kwargs)
        self.sex = 1
    
    def mortality_rate(self, minimum_mortality = 0.001):
        #the age thresholds for child_fn and elder_fn are guestimated from the data -- will need to check this when applying to other countries
        #minimum mortality is also guestimated from the data -- this is mostly to do away with a 0 mortality rate
        if self.age <=12:
            m = max(minimum_mortality, np.e**(-4.038012826363843 + -0.4620757529135555*self.age))
        elif self.age > 30:
            m = max(minimum_mortality, np.e**(-9.57843000677779 + 0.08987439511446728*self.age))
        else:
            m = minimum_mortality
        return m
    
    def update(self, dt):
        self.check_alive(dt)
        if self.isalive:
            self.age += dt
            self.check_fertility()
    
    
class Female(Individual):
    def __init__(self, *args, **kwargs):
        Individual.__init__(self, *args, **kwargs)
        self.n_children = 0 #keep track of n_children only on female side
        self.sex =0
        self.t_since_birth = None #period of time where she does not give birth again -- rounding it to 1 year-- this could probably be turned off tbh
        self.t_secondary_births =[]
        self.birth_age = []
        self.p_birth_array = [] #temporary container
        self.ismarried=False
        
        self.just_gave_birth = False #hack-y tracker to keep track of number of births per time unit for calibrating to Births/individual/year

    def mortality_rate(self, minimum_mortality = 0.001):
        if self.age <=12:
            m = max(minimum_mortality, np.e**(-4.154414473758413 + -0.4420609900131315*self.age)) #m = age specific mortality rate (deaths/individual/year)
        elif self.age > 30:
            m = max(minimum_mortality, np.e**(-10.527925730119318 + 0.10104736306154925*self.age))
        else: 
            m= minimum_mortality
        return m    

    def birth(self, dt):
        '''determines whether a woman gives birth. 
        other than imposing a 1 year refractory period births are 
        treated as independent of one another.. may want to change this'''
        if self.isfertile and self.ismarried:
            if self.t_since_birth != None:
                p_birth = np.max([0, self.base_p_child(dt) * self.birth_interval() * self.birth_desire()]) # prevent it from going negative
                
            else:
                p_birth = self.base_p_child(dt) #*self.birth_desire()
            p_birth = p_birth
            self.p_birth_array.append(p_birth)
            
            if np.random.random() <= p_birth:
                self.n_children +=1
                self.birth_age.append(self.age)
                if self.t_since_birth != None:
                    self.t_secondary_births.append((self.age, self.t_since_birth))
                self.t_since_birth = 0 
                self.just_gave_birth = True #flips on if birth happened
                return True
            else:
                self.just_gave_birth = False
                return False
        else: 
            self.p_birth_array.append(0)
            self.just_gave_birth = False
            return False
    
    def birth_interval(self):
        '''scaling factor that modifies p_child given t years after the previous birth'''
        weibull_exponentiated_params = { '15-19':  [7.008200010283388e-07, 0.18519982325420162, 4955315.919483488],
                                         '20-29': [0.617506311482934, 0.6131814950322227, 15.055481013015964],
                                         '30-39': [0.004432998576956875, 0.25617711748578326, 327.0658166547044],
                                         '40-49': [8.433674445639343e-10, 0.10913540296074727, 110510.2718509799]}
        if self.age >= 15 and self.age <20:
            p_birth_given_time = Weibull_exponentiated.cdf(weibull_exponentiated_params['15-19'], self.t_since_birth)
        elif self.age < 30:
            p_birth_given_time = Weibull_exponentiated.cdf(weibull_exponentiated_params['20-29'], self.t_since_birth)
        elif self.age < 40:
            p_birth_given_time = Weibull_exponentiated.cdf(weibull_exponentiated_params['30-39'], self.t_since_birth)
        elif self.age < 50:
            p_birth_given_time = Weibull_exponentiated.cdf(weibull_exponentiated_params['40-49'], self.t_since_birth)
        else:
            p_birth_given_time = 0
        
        return p_birth_given_time
            
    def birth_desire(self):#obtained from Bangladesh DHSS 2014 Table 6.1
        '''scaling factor that modifies p_child given they have n living children already
        Based on the percent distribution of women aged 15-49 who still desire children given n living children'''
        floor = self.base_fertility_params[-1]
        params = [ 1.36874821,  floor,  1-floor, -4.101415]
        p_desire = sigmoid(params, self.n_children) #this data is on still LIVING children...
        return p_desire

    def update(self, dt):
        self.check_alive(dt)
        if self.isalive:            
            if self.t_since_birth != None:
                self.t_since_birth += dt
            self.birth(dt)
            self.age += dt #advance time
            self.check_fertility() #advance time            
```

Calibrating Fertility

```python
class Children_Number:
    def __init__(self,n_children_file = 'data/Bangladesh_children_distribution_all_women_2014.txt', asfr_file = '../data/Bangladesh_DHSS_age_specific_fertility_2014.txt', scale = 1):
        self.raw_data = [_.strip().split('\t') for _ in open(n_children_file).readlines()]
        self.n_children = {}
        self.mle_params = {}
        self.f_children = {}
        self.smoothed_obs_births_per_f = {} #la place smoothed probabilities of each child count, split by age
        self.scale = scale
        self.observed_asfr = read_asfr_data(asfr_file)
        
        for line in self.raw_data[1:-1]:
            age_cohort = line[0]
            N_women = float(line[13].replace(',', ''))
            n_children_array = [round(float(x)/100. * N_women) for x in line[1:12]]
            f_children_array = [float(x)/100. for x in line[1:12]]
            self.n_children[age_cohort] = n_children_array
            self.f_children[age_cohort] = f_children_array
            self.smoothed_obs_births_per_f[age_cohort] = laplace_smoothing(n_children_array)
            
        self.age_cohorts = list(self.n_children.keys())
            
    def calculate_births_per_f(self, female_individuals):
        '''calculates the total births per individual female'''
        births_per_f = [i.n_children for i in female_individuals]
        births_per_f_counts = Counter(births_per_f)
        formatted_births_per_f_counts  = np.zeros(11) #format the counts so that it can be compared with the data
        #formatted_counts[i] = counts for group i, where i is a number
        
        for count, value in births_per_f_counts.items():
            if count < 10:
                formatted_births_per_f_counts[count] = value
            else: #10+ category
                formatted_births_per_f_counts[-1] += value
        return formatted_births_per_f_counts
        
    
    def simulate_births(self, params, return_additional_stats = False):
        individual_params = params[:6] #first 5 parameters are for the 4-d polynomial equation (a-e)
        f_married_param = params[6:] #final three is preference function and marriage function (f: n_children_preference_floor, g-i = marriage curve)
        
        female_individuals = [Female(0, individual_params, scale = self.scale) for _ in range(1000)]
        births_per_f_dict = {}
        asfr_dict = defaultdict(lambda: 0) #age specific fertility rate, split by age group
        
        #additional_stats
        f_married = [] #fraction_married
        secondary_birth_intervals = []
        
        n_just_gave_birth_dict = defaultdict(lambda :0)      
        age_cohort_rbounds = [float(x.split('-')[1]) for x in self.age_cohorts] #right bounds of agecohorts 3x: 15-19, hold only 19  
        tstep=0
        
        while tstep < 50: #because we are only interested in births, just run to the max fertility age
            
            #marriage logic
            p_marry = np.max([0,cdf_married(tstep) - cdf_married(tstep-1)])
            if p_marry > 1:
                p_marry = 1
            n_newly_married = np.random.binomial(len(female_individuals), p_marry) 
            unmarried_idxes = [idx for idx,f in enumerate(female_individuals) if f.ismarried==False]
            if len(unmarried_idxes) > 0:
                if len(unmarried_idxes) >= n_newly_married:
                    newly_married_idxes = np.random.choice(unmarried_idxes, n_newly_married, replace = False)
                else: 
                    newly_married_idxes = unmarried_idxes

                for idx in newly_married_idxes:
                    female_individuals[idx].ismarried=True   
            f_married.append(np.sum([f.ismarried for f in female_individuals]))
            
            #update females
            for i in female_individuals:
                i.update(1)
                if i.just_gave_birth:
                    n_just_gave_birth_dict[tstep] += 1
                                
            if tstep in age_cohort_rbounds:
                age_cohort = str(tstep-4) + '-' + str(tstep) #age cohorts are always binned in spans of 4
                births_per_f_dict[age_cohort] = self.calculate_births_per_f(female_individuals)
            tstep +=1
        
        for age_cohort in self.age_cohorts:
            lower,upper = int(age_cohort.split('-')[0]), int(age_cohort.split('-')[1])
            for age in range(lower, upper + 1): #range is not inclusive at the end, 15-20 yields 15..19
                asfr_dict[age_cohort] += n_just_gave_birth_dict[age] / ((upper-lower+1)*len(female_individuals)) * 1000.       
        
        
        if return_additional_stats: 
            for i in female_individuals:
                if len(i.t_secondary_births) > 1:
                    secondary_birth_intervals += i.t_secondary_births
            return births_per_f_dict, asfr_dict, f_married, secondary_birth_intervals
        else:
            return births_per_f_dict, asfr_dict
    
    
    def calculate_difference(self, params):
        '''calculates the KL divergence between the observed data and our simulation'''
        distance = 0
        sim_births_per_f_dict, sim_asfr_dict = self.simulate_births(params)
                                        
        for age_cohort in self.age_cohorts:         
            if age_cohort in self.age_cohorts[0:3]: #compare n_children/woman only for the first three age cohorts (15-30) because the n_children they have are most likely representative of current fertility trends
                sim_births_per_f = laplace_smoothing(sim_births_per_f_dict[age_cohort]) 
                distance += kl_divergence(self.smoothed_obs_births_per_f[age_cohort], sim_births_per_f)
        
        observed_asfr_array, simulated_asfr_array = [],[]
        for age_cohort in self.age_cohorts: #tbh i'm not sure this for loop is necessary, I just put it in to ensure that the right keys are being compared
            observed_asfr_array.append(float(self.observed_asfr[age_cohort]))
            simulated_asfr_array.append(float(sim_asfr_dict[age_cohort]))
        distance += kl_divergence(laplace_smoothing(observed_asfr_array), laplace_smoothing(simulated_asfr_array))
        
        return distance
    
    def calculate_distance_distribution(self, priors):
            '''generates a distance distribution (KL divergence technically) and determines the 30th, 10th, 5th percentiles'''
            distances = []
            for _ in range(1000):
                if _ % 100 == 0:
                    print('calculating distance distribution iteration {iteration} / 1000'.format(iteration = _))
                distance = N_Children.calculate_difference([priors[key]() for key in priors])
                distances.append(distance)
            plt.hist(distances)
            percentiles = [np.percentile(distances, 30), np.percentile(distances, 10), np.percentile(distances, 5), np.percentile(distances, 2.5)]
            return percentiles
    
    
    def sequential_pmc_ABC(self, n_particles=300, eps = None): 
        '''Population Montecarlo PMC algorithm designed to perform ABC and iteratively adapt the prior/proposal distribution to more closely resemble the posterior
        does this by iteratively reducing the tolerated error and progressively adapting the prior distribution
        follows the Algorithm 2 of Sadegh and vrugt 2014: Approximate Bayesian computation using Markov Chain Monte Carlo simulation: DREAM_ABC_ and the code at https://stackoverflow.com/questions/37496969/population-monte-carlo-implementation
        eps = initial error terms'''
        coefficients_mean = [-1.20e-06,  1.80e-04, -9.63e-03,  2.09e-01,-1.385] #4-d polynomial
        scales_std = [1e-8, 1e-6, 1e-10, 1e-7, 1e-1]
        priors = {'a': lambda: np.random.normal(coefficients_mean[0], scales_std[0]), 
                  'b': lambda: np.random.normal(coefficients_mean[1], scales_std[1]),
                  'c': lambda: np.random.normal(coefficients_mean[2], scales_std[2]),
                  'd': lambda: np.random.normal(coefficients_mean[3], scales_std[3]),
                  'e': lambda: np.random.normal(coefficients_mean[4], scales_std[4]),
                  'f': lambda: np.random.normal(0.4, 0.1),
                  
                  'g': lambda: np.random.normal(0.20978635, 0.02),
                  'h': lambda: np.random.normal(0.47, 0.05),
                  'i': lambda: np.random.normal(0.98, 0.005)}

                                                         
        def computeWeight(prevWeights,prevParticles,prevVariance,currentParticle):
            '''compute the weights needed to adapt the proposal distribution to be closer to the posterior'''
            #4D polynomial
            prior = np.prod([scipy.stats.norm.pdf(currentParticle, coefficients_mean[0], scales_std[0]), #x^4
                             scipy.stats.norm.pdf(currentParticle, coefficients_mean[1], scales_std[1]), #x^3
                             scipy.stats.norm.pdf(currentParticle, coefficients_mean[2], scales_std[2]), #x^2
                             scipy.stats.norm.pdf(currentParticle, coefficients_mean[3], scales_std[3]), #x
                             scipy.stats.norm.pdf(currentParticle, coefficients_mean[4], scales_std[4]),
                             scipy.stats.norm.pdf(currentParticle, 0.4, 0.1),
                             
                             scipy.stats.norm.pdf(currentParticle, 0.20978635, 0.03),
                             scipy.stats.norm.pdf(currentParticle, 0.47, 0.05),
                             scipy.stats.norm.pdf(currentParticle, 0.98, 0.005)])
            
            denom = 0.0
            scale = 1e5
            for i in range(len(prevParticles)):
                denom += prevWeights[i]  * scipy.stats.multivariate_normal(scale * currentParticle, 
                                                                           scale * prevVariance,
                                                                          allow_singular=True).pdf(scale * prevParticles[i])
            return prior/denom    
        
        if not eps:
            eps = self.calculate_distance_distribution(priors) #automatically calculate the 30th, 10th, 5th percentiles if not given
            print(eps)
        
        iterations = len(eps)
        distance = 1e9
        variance = {} #for multivariate, this is covariance matrix
        particles = defaultdict(list)
        weights = np.empty([iterations,n_particles])
        simData = np.empty(n_particles)
        
        
        for j in range(iterations):
            print('starting {j}th iteration'.format(j=j))
            if j == 0:
                for i in range(n_particles):
                    while(distance > eps[j]):
                        draw = [priors[key]() for key in priors]
                        distance = self.calculate_difference(draw)
                    if i in [0, 100, 200, n_particles - 1]:
                        print('particle {i} solution found'.format(i = i))
                    particles[j].append(draw)
                    weights[j][i] = 1./n_particles
                    distance = 1e9
                particles[j] = np.asarray(particles[j])
                variance[j] = 2 * np.cov(particles[j].T) #np.var(particles[j].T)
            else:
                for i in range(n_particles):
                    while(distance > eps[j]):
                        prelim_draw_index = np.random.choice(range(len(particles[j-1])), p = weights[j-1]) #draw a new parameter from the weighted distribution of previously identified parameters; random choice only works on 1-D arrays (no nested)
                        prelim_draw = particles[j-1][prelim_draw_index]
                        #draw = np.random.normal(prelim_draw, np.sqrt(variance[j-1])) #peturb the new prior in proportion to the stdev of the previous iteration
                        #distance = self.calculate_difference([draw])
                        draw = np.random.multivariate_normal(prelim_draw, variance[j-1]) #multivariate_normal accepts the covariance
                        distance = self.calculate_difference(draw)
                    if i in [0, 100, 200, n_particles - 1]:
                        print('particle {i} solution found'.format(i = i))

                    particles[j].append(draw)
                    weights[j][i] = computeWeight(weights[j-1],particles[j-1],variance[j-1],particles[j][i])
                    distance = 1e9
                particles[j] = np.asarray(particles[j])
                variance[j] = 2 * np.cov(particles[j].T) #np.var(particles[j].T)
                weights[j] = normalize(weights[j])
        self.posteriors = np.asarray(particles)
        self.max_posteriors = [np.mean(distribution) for distribution in self.posteriors.item()[j].T] 
        #at the end of the loop, j is the last iteration

    def plot(self, name):
        print(self.max_posteriors)
        
        sim_births_per_f_dict, sim_asfr_dict = self.simulate_births(self.max_posteriors) #[self.max_posterior])

        print(sim_births_per_f_dict)
        #plt.figure(figsize=(8,5))
        #plt.bar(self.observed_asfr.keys(), self.observed_asfr.values() , alpha = 0.5, color = 'blue', label = 'Bangladesh DHSS')
        #plt.bar(sim_asfr_dict.keys(), sim_asfr_dict.values(), alpha = 0.5, color = 'orange', label = 'simulation')
        
        plt.figure(figsize=(15,15))
        i = 1
        xticks = [x for x in range(10)] + ['10+']
        for age_cohort in sim_births_per_f_dict.keys():
            plt.subplot(3,3,i)
            plt.bar(range(len(self.n_children[age_cohort])), self.n_children[age_cohort] / np.sum(list(self.n_children[age_cohort])), color = 'blue', label = 'Bangladesh DHSS', alpha = 0.5)
            plt.bar(range(len(sim_births_per_f_dict[age_cohort])), sim_births_per_f_dict[age_cohort] / np.sum(list(sim_births_per_f_dict[age_cohort])), color = 'orange', label = 'simulation', alpha = 0.5)
            plt.xticks(range(len(xticks)), xticks)
            plt.title(age_cohort + ' n_children\n per woman')
            plt.xlabel('Years')
            i += 1
        plt.tight_layout()
        plt.legend()
        plt.savefig(name+ '.svg')
        
N_Children = Children_Number()
#N_Children.sequential_pmc_ABC()
#np.save('poly4d_v2_unweighted_only', N_Children.posteriors)
```

```python
class Birth_Intervals:
    distributions = {'weibull':Weibull,
                    'weibull_exponentiated': Weibull_exponentiated}
    
    def __init__(self):
        pass
    
    @classmethod
    def initiate(cls, file='../data/Bangladesh_birth_intervals_2014.txt'):
        B = Birth_Intervals()
        B.file = '../data/Bangladesh_birth_intervals_2014.txt'
        B.data = [_.strip().split(' ') for _ in open('../data/Bangladesh_birth_intervals_2014.txt').readlines() if '#' not in _][0:-1]
        B.p_age_cohorts = {}
        B.n_births = {} #n_births by age cohort (back-calculated)
        B.n_births_std_dev = {}
        B.xi_array = {}
        B.total_births = {} #total births per age cohort
        B.age_cohorts = []
        
        B.birth_intervals = ['7-17', '18-23', '24-35', '36-47', '48-59', '60+']
        B.birth_interval_midpoints = []
        for b in B.birth_intervals:
            if len(b.split('-')) > 1:
                start, stop = float(b.split('-')[0])/12., float(b.split('-')[1])/12.
                B.birth_interval_midpoints.append(np.mean([start, stop]))
        
        B.birth_interval_midpoints.append(5.)
        B.year_limits = [float(x.split('-')[0])/12. for x in B.birth_intervals[:-1]] + [5.]
        
        for line in B.data[1:]:
            B.age_cohorts.append(line[0])
            B.p_age_cohorts[line[0]] = np.asarray([float(x)/100. for x in line[1:7]])
            n_births = float(line[-2].replace(',', ''))
            B.total_births[line[0]] = n_births
            B.xi_array[line[0]] = np.asarray([float(x)/100. for x in line[1:7]]) * n_births 
        
        
        for age_cohort in B.age_cohorts:
            B.n_births[age_cohort] = B.total_births[age_cohort] * B.p_age_cohorts[age_cohort] #n_births by age cohort (back-calculated)
            B.n_births_std_dev[age_cohort] = np.sqrt(B.total_births[age_cohort] * B.p_age_cohorts[age_cohort] * (1. - B.p_age_cohorts[age_cohort]))
        B.optimize()
        B.calculate_simulated_values()
        return B
    
    def return_randomized(self):
        randomized_dict = defaultdict(list)
        for age_cohort in self.age_cohorts:
            for p in self.p_age_cohorts[age_cohort]:
                randomized_dict[age_cohort].append(np.random.binomial(self.total_births[age_cohort], p))
        return randomized_dict   
    
    def multinomial_birth_array_likelihood(self, distribution_type, params, age_cohort, data_dict, return_negative = False):
        '''assuming wait time distributions follows a weibull or Weibull_exponentiated distribution. 
        DHSS data is discretized into bins, hence the combination of multinomial distributions with the integral range of the Weibull/Weibull_expxonentiated distribution
        The x_array = the x_i for all i year categories where i = {1..5, 5+}'''
        x_array = data_dict[age_cohort]
        p_array = []
        for i,start in enumerate(self.year_limits):
            if i != len(self.year_limits) -1:
                stop = self.year_limits[i + 1]
                p_array.append(Birth_Intervals.distributions[distribution_type].integral(params, start, stop))
            else:
                p_array.append(1. - Birth_Intervals.distributions[distribution_type].cdf(params, 5.))
        log_p_i_product_sum = 0
        for x_i, p_i in zip(x_array, p_array):
            log_p_i_product_sum += (x_i * np.log(p_i))
            
        loglikelihood = log_p_i_product_sum 
        if not return_negative:
            return loglikelihood
        else:
            return -loglikelihood

    def optimize(self):
        n_iter = 1000
        self.mle_weibull_params = defaultdict(lambda: np.empty((n_iter, 2)))
        self.mle_weibull_exponentiated_params = defaultdict(lambda: np.empty((n_iter, 3)))
        for _ in range(n_iter):
            xi_array = self.return_randomized()
            if _ % 100 ==0:
                print(_)
                print(xi_array)                

            for age_cohort in self.age_cohorts:
                optimize_fn = lambda x0: self.multinomial_birth_array_likelihood('weibull', [x**2 for x in x0], age_cohort, xi_array, return_negative = True) #weibull distributions only defined when all parameters > 0
                mle_params = scipy.optimize.minimize(optimize_fn, [1,1])['x']
                self.mle_weibull_params[age_cohort][_]= [x**2 for x in mle_params]

                optimize_fn = lambda x0: self.multinomial_birth_array_likelihood('weibull_exponentiated', [x**2 for x in x0], age_cohort, xi_array, return_negative = True)
                mle_params = scipy.optimize.minimize(optimize_fn, [1,1,1])['x']
                self.mle_weibull_exponentiated_params[age_cohort][_]= [x**2 for x in mle_params]
        
        self.max_weibull_params = {}
        self.max_weibull_exponentiated_params = {}
        
        self.weibul_runs = defaultdict(lambda: np.empty((n_iter, 1000)))
        self.exp_weibul_runs = defaultdict(lambda: np.empty((n_iter, 1000)))
        x_array = np.linspace(0.0001, 10, 1000)
        for age_cohort in self.age_cohorts:
            self.max_weibull_params[age_cohort]  = np.median(self.mle_weibull_params[age_cohort], axis=0)
            self.max_weibull_exponentiated_params[age_cohort] = np.median(self.mle_weibull_exponentiated_params[age_cohort], axis=0)
            
            for _ in range(n_iter):
                self.weibul_runs[age_cohort][_] = [weibull_distribution(self.mle_weibull_params[age_cohort][_],x) for x in x_array]
                self.exp_weibul_runs[age_cohort][_] = [weibull_exponentiated_distribution(self.mle_weibull_exponentiated_params[age_cohort][_],x) for x in x_array]
                
            
    def calculate_simulated_values(self):
        self.weibull_sim_p_births, self.exp_weibull_sim_p_births = {}, {}
        self.weibull_sim_n_births, self.exp_weibull_sim_n_births = {}, {}
        self.weibull_sim_n_births_stdev, self.exp_weibull_sim_n_births_stdev = {}, {}
        for age_cohort in self.age_cohorts:
            weibul_sim_values = []
            weibul_exponentiated_sim_values = []
            for x in self.birth_intervals[:-1]:
                age_range = x.split('-')
                if len(age_range) > 1:
                    start, stop = int(age_range[0])/12, (int(age_range[1])+1)/12
                    weibul_sim_values.append(weibull_integral(self.max_weibull_params[age_cohort], start, stop))
                    weibul_exponentiated_sim_values.append(weibull_exponentiated_integral(self.max_weibull_exponentiated_params[age_cohort], start, stop))
                    
            weibul_sim_values.append(1.-weibull_cdf(self.max_weibull_params[age_cohort], 5)) #add the 60+ grouop
            weibul_exponentiated_sim_values.append(1. - weibull_exponentiated_cdf(self.max_weibull_exponentiated_params[age_cohort], 5))
            
            self.weibull_sim_p_births[age_cohort] = np.asarray(weibul_sim_values)
            self.exp_weibull_sim_p_births[age_cohort] = np.asarray(weibul_exponentiated_sim_values)
            
            self.weibull_sim_n_births[age_cohort] = self.weibull_sim_p_births[age_cohort] * self.total_births[age_cohort]
            self.exp_weibull_sim_n_births[age_cohort] = self.exp_weibull_sim_p_births[age_cohort] * self.total_births[age_cohort]
            
            self.weibull_sim_n_births_stdev[age_cohort] = np.sqrt(self.weibull_sim_p_births[age_cohort] * self.total_births[age_cohort] * 1- self.weibull_sim_p_births[age_cohort]) #sqrt(npq)
            self.exp_weibull_sim_n_births_stdev[age_cohort] = np.sqrt(self.exp_weibull_sim_p_births[age_cohort] * self.total_births[age_cohort] * 1- self.exp_weibull_sim_p_births[age_cohort]) #sqrt(npq)
            
    def plot_mle_distributions(self):
        i=1
        plt.figure(figsize=(10,10))
        x_array = np.linspace(0.0001, 10, 1000)
        for age_cohort in self.age_cohorts:
            plt.subplot(2,2,i)
            plt.plot(x_array, [weibull_distribution(self.max_weibull_params[age_cohort],x) for x in x_array], color = 'orange',
                    label ='weibull\n (Î» = {scale}, k = {shape})'.format(scale = round(self.max_weibull_params[age_cohort][0],2), shape = round(self.max_weibull_params[age_cohort][1], 2)))
            
            plt.plot(x_array, [weibull_exponentiated_distribution(self.max_weibull_exponentiated_params[age_cohort],x) for x in x_array], color = 'green',
                    label = 'exponentiated weibull\n (Î» = {scale}, k = {shape1}, Î± = {shape2})'.format(scale = round(self.max_weibull_exponentiated_params[age_cohort][0],2), 
                                                                                                                shape1 = round(self.max_weibull_exponentiated_params[age_cohort][1], 2),
                                                                                                               shape2 = round(self.max_weibull_exponentiated_params[age_cohort][2],2)))
            
            plt.fill_between(x_array, np.percentile(self.weibul_runs[age_cohort], 2.5, axis = 0), np.percentile(self.weibul_runs[age_cohort], 97.5, axis = 0), color = 'orange', alpha = 0.3)
            plt.fill_between(x_array, np.percentile(self.exp_weibul_runs[age_cohort], 2.5, axis = 0), np.percentile(self.exp_weibul_runs[age_cohort], 97.5, axis = 0), color = 'green', alpha = 0.3)
            plt.ylim(0,0.8)
    
    
            plt.title('Ages ' + age_cohort + ' Birth Intervals')
            plt.legend()
            i += 1
            
            plt.xlabel('Time (years)')
            plt.ylabel('Proportion')
            plt.ylim(0,0.7)
        plt.tight_layout() 
        #plt.savefig('Birth_interval_weibulls.svg')
        plt.savefig('Birth_interval_weibulls.png', dpi = 300)
        
    def plot_data_comparison(self):
        plt.figure(figsize=(10,10))
        for i, age_cohort in enumerate(self.age_cohorts):
            plt.subplot(2,2,i +1)
            
            plt.bar([x for x in range(1,7)], self.n_births[age_cohort], 
                         yerr = self.n_births_std_dev[age_cohort],
                         color = 'blue', label = 'Bangladesh DHSS', width = 0.25, ecolor='grey')
            
            plt.bar(np.asarray([x for x in range(1,7)]) + 0.25, self.weibull_sim_n_births[age_cohort], 
                    yerr = self.weibull_sim_n_births_stdev[age_cohort],
                    color = 'orange', label = 'Weibull', width = 0.25, ecolor='grey' )
            
            plt.bar(np.asarray([x for x in range(1,7)]) + 0.5, self.exp_weibull_sim_n_births[age_cohort], 
                    yerr = self.exp_weibull_sim_n_births_stdev[age_cohort],
                    color = 'green', label = 'Exponentiated Weibull', width = 0.25, ecolor='grey')

            plt.xticks(np.asarray(range(1,7)) + 0.2, self.birth_intervals)
            plt.title('Ages ' + age_cohort + ' Birth Intervals')
            plt.legend()
            plt.ylabel('Births')
            plt.xlabel('Age Groups')
        plt.tight_layout()
        #plt.savefig('Birth_interval_comparison.svg')
        #plt.savefig('Birth_interval_comparison.png', dpi = 300)
import warnings
warnings.filterwarnings("ignore") #runtime errors of zero because it occasionally chooses parameters with distributions like [0,0,0,0,0,1] -- dunno how to fix...so I'm suppressing it
B= Birth_Intervals.initiate()
```

```python
from scipy.optimize import curve_fit 
from scipy.misc import derivative

prop_married = 1. - np.asarray([1-0.163, 0.548, 0.155, 0.041, 0.012, 0.008, 0.006, 0.002]) #Bangladesh 2014 DHSS Report (Table 4.3)
age_groups = ['15-19', '20-24', '25-29', '30-34', '35-39', '40-44', '45-49']
n_married = np.asarray([4485, 4485, 3816, 3534, 3084, 2334, 2105, 1769])
stdev_married = (prop_married * (1-prop_married)) / n_married


midpoint_ages = np.asarray([15] + [np.mean([float(x.split('-')[0]), float(x.split('-')[1])]) for x in age_groups])
shifted_midpoints = midpoint_ages - midpoint_ages[0]

randomized_params = np.empty((100, 3))
randomized_curves = np.empty((100, 1000))
xp = np.linspace(0, 100, 1000)
def growth_rate(x, p0,  r, k):
    '''population growth logistic equation: Verhulst equation'''
    return (k*p0*np.exp(r*x)) / (k + p0*(np.exp(r*x) - 1))

for _ in range(100):
    randomized_sample = []
    for n,p in zip(n_married, prop_married):
        randomized_sample.append(np.random.binomial(n,p)/n)
        
    params, cov = curve_fit(growth_rate, midpoint_ages, randomized_sample)
    
    randomized_params[_] = params
    sim_values=growth_rate(xp, *list(randomized_params[_]))
    randomized_curves[_] = sim_values
    
    

def cdf_married(age, coefficients = [0.20978635, 0.47889714, 0.98557189]):
    def growth_rate(x, p0,  r, k):
        '''population growth logistic equation: Verhulst equation'''
        return (k*p0*np.exp(r*x)) / (k + p0*(np.exp(r*x) - 1))
    
    x = age - 15 #shift the growth_rate equation for better calculation
    y = growth_rate(x, *coefficients)
    return y

def pdf_married(age, coefficients = [0.20978635, 0.47889714, 0.98557189]):
    '''the derivative of the cdf. Quotient rule notation = T'B - TB' / B^2'''
    p0, r, k = coefficients[0], coefficients[1], coefficients[2]
    x = age - 15
    T_prime = r * k * p0 * np.exp(r*x)
    B_prime = r * p0 * np.exp(r*x)
    T = (k*p0*np.exp(r*x))
    B = (k + p0*(np.exp(r*x) - 1))
    return (T_prime*B - T*B_prime) / (B**2)
                                     
plt.errorbar(midpoint_ages, prop_married, yerr=stdev_married, fmt = 'o', ms = 3, label = 'Bangladesh DHSS')
plt.plot(xp, cdf_married(np.linspace(0,100,1000)), color = 'orange', label = 'Logistic growth curve\n (p0 = {p0}, r = {r}, k = {k})'.format(p0 = 0.21, r = 0.48, k = 0.99))
plt.fill_between(xp, np.percentile(randomized_curves, 2.5, axis = 0), np.percentile(randomized_curves, 97.5, axis = 0), color = 'orange', alpha = 0.3)

plt.legend()
plt.xlabel('Age')
plt.ylabel('Proportion')
plt.title('Proportion Individuals Married')
plt.xlim(14,30)
#plt.savefig('marriage_cdf.svg')
```
